```python
beer=pd.read_csv('data/beer.csv')
```

```python
beer.plot(kind='box',y='abv')
```

## 25%지점과 75%지점이 어딘지 알아보고 이상점 없애기

```python
beer['abv'].describe()
```

```python
q1=beer['abv'].quantile(0.25)
```

```python
q3=beer['abv'].quantile(0.75)
```

```python
iqr=q3-q1
```

```python
condition=(beer['abv']<q1-1.5*iqr)|(beer['abv']>q3+1.5*iqr) # 25% 지점에서 1.5*iqr 미만이거나 75% 지점에서 1.5*iqr 이상인 이상점 찾기
```

```python
beer[condition] # 참이슬과 그레이 구스 보드카는 맥주보다 도수가 훨씬 높아서 이상점, 세번째는 맥주인데 0.055가 잘못 기입돼서 이상점
```

```python
beer.loc[2250,'abv']=0.055 # 이상점 수정하기
beer.drop(beer[condition].index,inplace=True) # 이상점 없애기, condition 인덱스에 있는 값들이 삭제됨.
```

```python
condition=(beer['abv']<q1-1.5*iqr)|(beer['abv']>q3+1.5*iqr)
beer[condition] 
```

```python
beer.plot(kind='box',y='abv')
```

*이상점이 모두 없어짐!!*

```python
exam=pd.read_csv('data/exam_outlier.csv')
```

```python
exam.plot(kind='scatter',x='reading score',y='writing score') # wirting 점수가 비정상적으로 높은 이상점 발견
```

```python
exam.corr() # 쓰기와 읽기 점수의 상관관계도 이상점 때문에 생각보다 작게 나옴.
```

```python
exam[exam['writing score']>100]
```

```python
exam.drop(51,inplace=True)
```

```python
exam.corr() # 읽기와 쓰기 점수의 상관관계가 아주 높아짐.
```

```python
condition1=(exam['writing score']>90)&(exam['reading score']<40)
exam[condition1] # 분석하는데 방해가 될 만한 요소임. 지우자.
```

```python
a=covid[condition].index
a
```
