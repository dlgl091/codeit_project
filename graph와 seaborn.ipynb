```python
sns.kdeplot(df4['Height'],bw=0.5) 
```

**seaborn의 kde는 울퉁불퉁한 선들을 매끄러운 곡선으로 만들어줌(확률밀도함수로 만들어줌, 실제 분포와는 다름. 추측을 해서 매끄럽게 만든 것이기 때문)**

**bw를 이용해서 추측의 정도를 조절할 수 있음**

```python
sns.distplot(df4['Height'],bins=15) # 히스토그램 위에 kde 곡선을 그림. 관측된 데이터들로부터 히스토그램을 구한 후 구해진 히스토그램을 정규화하여 확률밀도함수로 사용하는 것이다.
```

```python
sns.violinplot(y=df4['Height']) # kde 그래프를 대칭되게 그린 그래프. 바이올린처럼 생겼다고 violinplot
```

```python
sns.kdeplot(df4['Height'],df4['Weight']) # x축과 y축 값을 3차원으로 그린 그래프. 선이 가까울수록 가파름, 멀수록 평평함. 그래프의 가로축, 세로축 상상해보고 가파름과 평평함으로 값들 분석
```

```python
df4.plot(kind='scatter',x='Height',y='Weight')
```

```python
sns.lmplot(data=df4,x='Height',y='Weight') # 가운데는 회귀선. 키가 175인 사람이 대충 72kg이 나갈 것이라고 예측할 수 있음. 정확x 주변 점들의 위치를 보고 x와 y의 상관관계를 알 수 있음
```

```python
laptop_df['brand_nation']=laptop_df['brand'].map(brand_nation) # 'brand'값들을 모두 brand_nation의 값으로 바꿔줌.
laptop_df
```

**응용:** laptop_df.insert(loc=1,column='brand_nation',value=laptop_df['brand'].map(brand_nation)) ->brand_nation의 위치를 선정해서 열 추가

```python
nation_groups=laptop_df.groupby('brand_nation')
type(nation_groups) # DataFrameGroupBy라는 데이터타입임.
```

```python
nation_groups.count()
```
